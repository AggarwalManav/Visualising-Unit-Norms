# Unit Norm 

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from numpy.linalg import norm
%matplotlib inline
```

```python
def norm_calc(v,p):
    if p==np.inf:
        return max(v)
    v=abs(v)
    v=v**p
    norm=v.sum()
    norm=norm**(1/p)
    return norm
```

## L-1 Norm 

```python
a=np.array([1,2])
norm(a,1)
```

## L-2 Norm 

```python
norm(a,2)
```

## L-0 Norm 

```python
norm(a,0)
```

## L-0.5 Norm 

```python
norm(a,0.5)
```

## L-max(infinity) Norm

```python
norm(a,np.inf)
```

# Plotting L-p Norms 

```python
x=np.linspace(-40,40,1000)
y=x
X,Y=np.meshgrid(x,y)
```

```python
p=[0.5,1,2,np.inf]
```

# L-1 Norm 

```python
Z=X+Y
```

```python
ax=plt.axes(projection="3d")
ax.plot_surface(X,Y,Z,cmap="plasma")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.set_title("L1 Norm")
plt.savefig("L1norm.png",dpi=600)
```

## L-2 Norm 

```python
Z=(X**2+Y**2)**0.5
ax=plt.axes(projection="3d")
ax.plot_surface(X,Y,Z,cmap="plasma")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.set_title("L2 Norm")
plt.savefig("L2norm.png",dpi=600)
```

## L-0.5 Norm

```python
x_new=np.linspace(1,40,1000)
y_new=x_new
X_new,Y_new=np.meshgrid(x_new,y_new)
Z=(X_new**0.5+Y_new**0.5)**2
ax=plt.axes(projection="3d")
ax.plot_surface(X_new,Y_new,Z,cmap="plasma")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.set_title("L-0.5 Norm")
plt.savefig("LHalfnorm.png",dpi=600)

```

## L-max Norm 

```python
Z=np.zeros([1000,1000])
for i in range(1000):
    for j in range(1000):
        Z[i,j]=max(X[i,j],Y[i,j])
ax=plt.axes(projection="3d")
ax.plot_surface(X,Y,Z,cmap="plasma")
ax.view_init(30,60)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.set_title("L-max Norm")
plt.savefig("LMaxnorm.png",dpi=600)
```

## L-0 Norm 

```python
Z=np.zeros([1000,1000])
for i in range(1000):
    for j in range(1000):
        if X[i,j]==0 and Y[i,j]==0:
            Z[i,j]=0
        elif X[i,j]!=0 and Y[i,j]!=0:
            Z[i,j]=2
        else:
            Z[i,j]=1
ax=plt.axes(projection="3d")
ax.plot_surface(X,Y,Z,cmap="plasma")
ax.view_init(30,60)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.set_title("L-0 Norm")
plt.savefig("LZeronorm.png",dpi=600)
```
